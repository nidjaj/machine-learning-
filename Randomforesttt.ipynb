```
import pandas as pd
```

```
df=pd.read_csv("http://13.234.66.67/summer19/datasets/social.csv")
```

```
 # schema 
  df.info()
```

```
# top five rows
df.head()
```

```
# removing user id
filterdata=df.iloc[0:,2:].values
```

```
# now we can apply training testing
features=filterdata[0:,0:2]
```

```
# label finding
label=filterdata[0:,2]
```

```
# training and testing
from sklearn.model_selection import train_test_split
```

```
X,x,Y,y=train_test_split(features,label,test_size=0.2)
```

```
# now features scaling 
from sklearn.preprocessing import StandardScaler
```

```
sc=StandardScaler()
```

```
# Apply scalling in train data only
X=sc.fit_transform(X)
```

```
X
```

```
# testing data scaling

x=sc.transform(x)
```

```
# KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```
# random forest with number of decision tree
rclf=RandomForestClassifier(n_estimators=20)
```

```
# now training
trained=rclf.fit(X,Y)
```

```
# now predicting
output=trained.predict(x)
```
